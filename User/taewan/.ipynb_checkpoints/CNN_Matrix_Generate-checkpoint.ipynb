```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot

###189*198###
file_directory = "/Users/taewan/Desktop/"
geneset = pd.read_csv(file_directory+"geneset.csv")
genedata = pd.read_csv(file_directory+"genesymbols.csv")


testgene = np.random.randint(0,10,size =(2,len(genedata.columns)))
testgene = pd.DataFrame(testgene , columns=genedata.columns)

###Making Matrix###
matrix = np.zeros((len(testgene), len(geneset), 198))
print(len(testgene) , len(geneset))
###Each GeneSet###
geneset['gene_symbols'][188].split()

fig = pyplot.figure(2)

cmap2 = mpl.colors.LinearSegmentedColormap.from_list('my_colormap',
                                           ['white','red'],
                                           256)

for i in range(len(testgene)):
    for j in range(len(geneset)):
        gene_symbols = geneset['gene_symbols'][j].split(",")
        cnt = 0 
        for k in range(len(gene_symbols)):
            if gene_symbols[k] in testgene.columns :
                matrix[i][j][cnt] = testgene[gene_symbols[k]][i]
                cnt+=1
    
    img2 = pyplot.imshow(matrix[i],interpolation='nearest',
                    cmap = cmap2,
                    origin='lower')
    if(i == 0 ):
        pyplot.colorbar(img2,cmap=cmap2)
    fig.savefig(file_directory+"image" +str(i) +".png" )
#     pyplot.show()
```

```python
cmap = mpl.colors.ListedColormap(['black','red'])
bounds=[0,5,10]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

# tell imshow about color map so that only set colors are used
img = pyplot.imshow(matrix[0],interpolation='nearest',
                    cmap = cmap,norm=norm)

# make a color bar2
pyplot.colorbar(img,cmap=cmap,
                norm=norm,boundaries=bounds,ticks=[-5,0,5])

pyplot.show()
```
